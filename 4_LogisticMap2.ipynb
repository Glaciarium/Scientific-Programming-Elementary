# Logistic Map 相关补充

   逻辑斯蒂映射 （Logistic Map），又称单峰映象，是一个二次多项式映射(递归关系)，经常作为典型范例来说明复杂的混沌现象是如何从非常简单的非线性动力学方程中产生的。生物学家 Robert May 在1976年的一篇论文中推广了这一映射，它在一定程度上是一个时间离散的人口统计模型，类似于 Pierre Francois Verhulst 首次提出的方程。它的数学形式(迭代方程）为：
$$ x_{n+1}=rx_n(1-x_n) $$

当 $r$ 超过 $[1,4]$ 时，就会发生混沌现象。该非线性差分方程意在观察两种情形:  
-当人口规模很小时，人口将以与当前人口成比例的速度增长进行繁殖。  
-饥饿(与密度有关的死亡率) ，其增长率将以与环境的”承受能力”减去当前人口所得值成正比的速度下降

   然而，Logistic Map作为一种人口统计模型，存在着一些初始条件和参数值(如 $r>4$ )为某值时所导致的混沌问题。这个问题在较老的瑞克模型中没有出现（该模型也展示了混沌动力学）。

   同时，由于该模型可以用来模拟生物种群的生长行为，所以Logistic映射也叫“虫口模型”。其中 $r$ (分支参数，有时也用 $\mu$ 表示）也可以解释为在t时刻种群占最大可能种群规模的比例。我们将原方程变形为： 
$$x(t+1)-x(t)=(r-1) x(t)-rx(t)^2$$
    其中，$x(t+1)$可以解释为种群的生长率（即一个单位周期内，种群数量的变化）,$(r-1) x(t)$可以解释为虫种群的出生，$rx(t)^2$为种群的消亡。其中消亡项和$x(t)^2$有关，也就是说种群数量越多，消亡得越快，这体现为该种群内部由于资源有限而引起的竞争。

# 李雅普诺夫指数 （Lyapunov Exponent)

- 定义  
    在数学领域中，**李雅普诺夫指数**（**Lyapunov Exponent**）或**李雅普诺夫特征指数**（**Lyapunov Characteristic Exponent**） 用于量化动力系统中无限接近的轨迹之间的分离率。具体而言，相空间中初始间隔$\delta Z_0$的两条轨迹的分离率为（假定分离可按线性近似来处理）
    $$|\delta Z( t )| ≈ e^{\lambda t} |\delta Z_0|$$  
    即$$\lambda = \dfrac{1}{n} ln|\dfrac{\delta x_n}{\delta x_0}|$$  
    其中$\lambda$即为李雅普诺夫指数。  
    
    当初始分离向量的方向不同时，其分离率也不同。因而存在李雅普诺夫指数谱（Spectrum of Lyapunov Exponents），其数量与相空间的维度相同。通常将其中最大的称为**最大李雅普诺夫指数**（**Maximal Lyapunov Exponent**，简称**MLE**），因为它决定了动力系统的可预测性。与矩阵的最大特征值一样，最大的李雅普诺夫指数决定了系统的主导行为,为了减小计算量，则只需计算最大李雅普诺夫指数。
    `李雅普诺夫指数是定量描述混沌运动的高度敏感性这一现象的量。`

_P.S._     
    - 负李雅普诺夫指数与耗散系统相关（表明体系的相体积在该方向上是收缩的，周期性的，即此方向的运动是稳定的）；    
    - 等于零的李雅普诺夫指数与保守系统相关；    
    - 正李雅普诺夫指数与混沌系统相关（表明体系的相体积在该方向上不断膨胀和折叠，以致吸引子中本来邻近的轨线变得越来越不相关，从而使初态对于任何不确定性的系统的长期行为成为不可预测，即所谓的**初值敏感性**）。
- 用Logistic Map产生的模拟时间序列数据，可以采用两种从实验数据时间序列回复动力学的方法，以计算混沌吸引子的Lyapunov Exponent：
    1. S.J.Chang和J.Wright的混合嫡法  
      仅适用于可化为一维凸映射的情况，一般来说总点数越大，所得结果越精确。
    2. A.Wolf的重构吸引子法  
      可以推广到相空间维数和动力学规律均不知道的更普遍的试验系统，原则上可以计算系统的全部正李雅普诺指数谱；一般适合嵌入维$m>1$的重构吸引子的时间序列。

## 方法一
$$\lambda = \dfrac{1}{n} ln|\dfrac{\delta x_n}{\delta x_0}|$$

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

r = np.linspace(0, 4.0, 10000)
x0 = 1 * np.ones(10000)
x1 = (1+0.01) * np.ones(10000)
iterations = 10000
lyapunov1 = np.zeros(10000)

def logistic(r,x):
    return r*x*(1-x)

for i in range(10000):
    x0 = logistic(r, x0)
    x1 = logistic(r, x1)  
    delta = x1 - x0
    lyapunov1 += np.log(abs(delta/0.01))
    # np.log()默认为自然对数；np.log10()为常用对数
    
fig,ax = plt.subplots(figsize=(20, 10))
# Horizontal line.
ax.axhline(color='k', lw=0.5, alpha=0.5)
# Negative Lyapunov Exponent.
ax.plot(r[lyapunov1 < 0],
         lyapunov1[lyapunov1 < 0] / iterations,
         '.k', ms=1)
# Positive Lyapunov Exponent.
ax.plot(r[lyapunov1 >= 0],
         lyapunov1[lyapunov1 >= 0] / iterations,
         '.r', ms=1)

ax.set_xlim(0, 4)
ax.set_ylim(-2, 1)
ax.set_xlabel('r')
ax.set_title("Lyapunov Exponent 1")
```

## 方法二 
$$\lambda = \dfrac{1}{n} \sum_{i=0}^{n-1}ln|f'(x_i)|$$

```python
fig,ax = plt.subplots(figsize=(20, 10))

r = np.linspace(1e-5, 4.0, 10000)
lyapunov2 = np.zeros(10000)
x = 1e-5 * np.ones(10000)

iterations = 10000

for i in range(10000):
    x = logistic(r, x)
    lyapunov2 += np.log(abs(r - 2 * r * x))

ax.axhline(color='k', lw=0.5, alpha=0.5)
ax.plot(r[lyapunov2 < 0],
         lyapunov2[lyapunov2 < 0] / iterations,
         '.k', ms=0.5)
ax.plot(r[lyapunov2 >= 0],
         lyapunov2[lyapunov2 >= 0] / iterations,
         '.r', ms=0.5)

ax.set_xlim(0, 4)
ax.set_ylim(-2, 1)
ax.set_xlabel('r')
ax.set_title("Lyapunov Exponent 2")
```

从分岔图知$r < 3$不动点的存在性，然后是2平衡点和4平衡点，以及当r属于参数空间的某些区域时的混沌行为。同时对应观察到李雅普诺夫指数的一个重要性质:当系统是混沌的时候李雅普诺夫指数指数是正的（图中红色部分）。[\[9\]](https://ipython-books.github.io/121-plotting-the-bifurcation-diagram-of-a-chaotic-dynamical-system/)

```python
import matplotlib.pyplot as plt
import numpy as np
fig,ax = plt.subplots(figsize=(20, 10))
r = np.linspace(3.9055,3.907,10000)
lyapunov2 = np.zeros(10000)
x = 1e-5 * np.ones(10000)
for i in np.arange(0,10000,0.1):
    x = logistic(r, x)
    lyapunov2 += np.log(abs(r - 2 * r * x))
    ax.axhline(color='k', lw=0.5, alpha=0.5)
ax.plot(r[lyapunov2 < 0],
         lyapunov2[lyapunov2 < 0] / iterations,
         '.k', ms=0.2)
ax.plot(r[lyapunov2 >= 0],
         lyapunov2[lyapunov2 >= 0] / iterations,
         '.r', ms=0.2)
ax.set_xlim(3.9055, 3.907)
ax.set_title("Lyapunov Exponent 2.1")
plt.show()
```
