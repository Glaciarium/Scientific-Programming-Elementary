# 1. 一些历史背景

## 1.1 爱德华·洛伦兹 _Edward Norton Lorenz_ 的研究历程

1938年，洛伦兹毕业于达特茅斯学院，获得数学本科学位；后来获得哈佛大学硕士学位（1940 年）和麻省理工学院博士学位（1948年）；1941-42学年在哈佛大学教授数学。同年他被安排进入美国陆军团担任天气预报员。1946年，洛伦兹在麻省理工学院（MIT）开始了气象学进修。他受到了著名的 MIT 气象学家_Carl-Gustav Rossby_ (1898-1957) 和 _Vilhelm Bjerknes_ (1862-1951) 的影响——他们的工作强调对大气运动的物理动力学解释，并提出了第一个理论和实践气象学的系统模型，这种天气预报系统被美国气象局和军方广泛使用于二战期间。  

作为哈佛大学的访问科学家，1950年洛伦兹的研究集中在大气的能量平衡上。后来，这项关于动态和统计系统理论基础的研究成为加州大学洛杉矶分校和麻省理工学院在统计天气预报（“MIT Statistical Forecasting Project”）方面联合研究工作的一部分。1959年预报项目结束时，洛伦兹被提升为副教授，他的研究开始专注于使用计算机建模的天气预报。他起初开发了一个模型，其中包含12个方程，并于1961年开始了一个使用计算机模拟天气模式的项目。由于内存不足，计算机无法创建复杂的模式，但它能够显示主要气象事件之间的相互作用，如龙卷风和飓风。在这个过程中，洛伦兹开始看到模式的出现，并能够在一定程度上准确地预测会出现什么样的天气模式，这符合当时学界普遍认知。<a href="https://www.encyclopedia.com/science/encyclopedias-almanacs-transcripts-and-maps/edward-norton-lorenz" title="Edward Norton Lorenz - Encyclopdia">[Edward Norton Lorenz - Encyclopdia]</a>    
           
同年冬天，洛伦兹正在使用一台简单的数字计算机 Royal McBee LGP-30 来运行他的天气模拟方程。当他将其跟踪到计算机打印输出时出现了奇怪的现象：由于计算机以6位精度工作，但打印输出将变量四舍五入为3位数字，因此像 0.506127 这样的值会被打印为 0.506 。根据当时的共识是，<span style="border-bottom:2px dashed gold;">如此微小的偏差不应该有实际效果（即结果偏差也不应太大）</span>。然而，洛伦兹发现初始条件的微小变化导致了长期结果的巨大变化，也就是说**天气系统的长期预测是不可能的**。洛伦兹将此称为对**初始条件的敏感依赖**——今天称为**混沌**。敏感依赖是混沌系统非线性的直接结果。在线性系统中，初始状态的变化会导致后续状态的成比例变化。像大气这样的非线性系统表现得不太好。这些系统中的变量相互耦合，因此变化会导致复杂的累积馈效应（例如，温度升高会增加湿度，这会导致更多的太阳辐射被吸收，从而进一步升高温度，从而增加湿度等）。

随后，洛伦兹更详细地研究了非线性方程的行为。他特别研究了这样一个系统的变化（后来被命名为**洛伦兹吸引子**）。洛伦兹发现的吸引子后来被证明是第一个**奇异吸引子**。与其他吸引子一样，非线性系统将被奇异吸引子吸引到一组状态。然而，一旦系统处于奇怪的吸引子中，附近的状态就会以指数级的速度相互发散，从而无法进行远程预测。然而，奇异吸引子是具有确定结构的稳定几何物体。 这为非线性系统的明显随机行为提供了顺序。例如，天气奇怪吸引子的潜在顺序——气候——解释了为什么类似的天气模式一遍又一遍地发生。[\[Edward Lorenz's Groundbreaking Study of Weather Patterns Leads in Part to the Development of Chaotic Dynamics - Encyclopdia\]](https://www.encyclopedia.com/science/encyclopedias-almanacs-transcripts-and-maps/edward-lorenzs-groundbreaking-study-weather-patterns-leads-part-development-chaotic-dynamics)    

## 1.2 洛伦兹方程 _Lorenz equations_

1963年，_Edward Lorenz_ 在 _Ellen Fetter_ 的帮助下，将原来描述大气对流的数学模型简化，得到了一组由三个常微分方程组成的系统，现在称为洛伦茨方程：

$$
\dfrac{dx}{dt} = \sigma (y - x) \\
\dfrac{dy}{dt} = x(\rho - z) - y \\
\dfrac{dz}{dt} = xy - \beta z \\
$$

其中，变量 $x$ 与对流速率成正比， $y$ 代表水平温度变化 , $z$ 代表数值温度变化。  
&emsp;&emsp;&emsp;常数 $\sigma$ , $\rho$ 和 $\beta$ 是系统参数：$\sigma$ 代表普朗特数 _Prandtl number_ , $\rho$ 代表瑞利数 _Rayleigh number_ , $\beta$ 代表辊宽比 _aspect ratio of the rolls_ ；  
&emsp;&emsp;&emsp;通常 $\sigma$ = 10，$\beta$ = 8/3，$\rho$ 不定。

### 基本性质  
- 解的对称性
- 耗散系统
- 不动点条件  
$$
\dot{x} = 0 \\
\dot{y} = 0 \\ 
\dot{z} = 0 \\
$$


## 1.3 洛伦兹吸引子 _Lorenz attractor_

对于吸引子，学术上并没有完善的定义，只能说吸引子（ _Attractor_ ）是微积分和系统科学论中的一个概念。**如果一个系统有朝某个稳态发展的趋势，这个稳态就叫做吸引子。**

吸引子分为平庸吸引子( _Mediocre attractor_ )和奇异吸引子（ _Strange attractor_ ）。例如一个钟摆系统，它有一个平庸吸引子，这个吸引子使钟摆系统向停止晃动的稳态发展。平庸吸引子有不动点（平衡）、极限环（周期运动）和整数维环面（概周期运动）三种模式。而不属于平庸的吸引子的都称为奇异吸引子——也就是说，如果一个吸引子具有分形结构，则它可以被称为奇异吸引子。奇异吸引子包括 _double-scroll_ 吸引子、_Hénon_ 吸引子、 _Rössler_ 吸引子和 _Lorenz_ 吸引子。一个具有混沌吸子的动态系统在局域是不稳定的，然而广域来看却可以是稳定的，因为这些动态点再怎么彼此分离，都不会离开吸子。  

2001年，沃里克·塔克尔（Warwick Tucker）证明出在一组确定的参数下，洛伦兹系统会表现出混沌行为，即显示出人们今天所知的奇异吸引子。

_P.S._ 彼此关系？  
洛伦茨振子（ _Lorenz oscillator_ ）能产生混沌流的三维动力系统，又称作洛伦茨系统。从技术角度看来，洛伦茨振子具有非线性、三维性和确定性。特别地，洛伦兹吸引子是洛伦兹系统的混沌解的集，也是是洛伦茨振子的长期行为对应的分形结构。

# 2. 探索 _Lorentz_ 系统在不同 r 值下动力学表现

## 2.1 分别画出 x、y、 z 随时间的演化图，并描述其特征。

```python
from scipy.integrate import odeint  
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Lorenz方程，用于求 L=[x,y,z] 点的导数 dL/dt
def lorenz(L,t,sigma,rho,beta):  
    x, y, z = L  # Location（L） = [x,y,z]
    dx_dt = sigma*(y-x) 
    dy_dt = x*(rho-z) - y  
    dz_dt = x*y - beta*z  
    return np.array([dx_dt,dy_dt,dz_dt])

t = np.arange(0.0, 40.0, 0.01) # 创建时间点 (start,stop,step)
paras = (10.0, 28.0, 8.0/3.0)  # 设置 Lorenz 方程中的参数 (sigma,rho,beta)


# 调用ode对lorenz进行求解, 用两个不同的初始值 L1、L2 分别求解
# odeint() 函数中定义导数函数的标准形式是 f(y,t) ，对于微分方程组y表示向量.
L1 = [1.0, 1.0, 1.0]  # 定义初值为 L1
track1 = odeint(lorenz, L1, t, args=paras) # 通过 paras 传递导数函数的参数
L2 = [1.0, 1.0, 1.0]  # 定义初值为 L2
track2 = odeint(lorenz, L1, t, args=(10.0, 10.0, 8.0/3.0))  # args 设置导数函数的参数
# 注意通过 args=paras 或 args = (10.0,28.0,8.0/3.0) 将参数 (sigma,rho,beta) 传递给导数函数 lorenz(L,t,sigma,rho,beta)。
# 参数 (sigma,rho,beta) 当然也可以不作为函数参数传递，而是在导数函数 lorenz() 中直接设置。但通过参数传递方法，使导数函数结构清晰、更为通用。另外，对于可变参数问题，使用这种参数传递方式就非常方便。

fig, ax1 = plt.subplots(figsize=(20, 10))
y1 = track1[:, 0]
y2 = track2[:, 0]
ax1.plot(t,y1,label="rho=28.0")
ax1.plot(t,y2,label="rho=10.0",color='pink')
ax1.set_title('x-t progress')

fig, ax2 = plt.subplots(figsize=(20, 10))
y1 = track1[:, 1]
y2 = track2[:, 1]
ax2.plot(t,y1,label="rho=28.0")
ax2.plot(t,y2,label="rho=10.0",color='pink')
ax2.set_title('y-t progress')

fig, ax3 = plt.subplots(figsize=(20, 10))
y1 = track1[:, 2]
y2 = track2[:, 2]
ax3.plot(t,y1,label="rho=28.0")
ax3.plot(t,y2,label="rho=10.0",color='pink')
ax3.set_title('z-t progress')
```

## 2.2 画三维图，描述系统在不动点附近如何运动。

```python
from mpl_toolkits.mplot3d import Axes3D

L3 = [1.0, 1.01, 1.0]
track3 = odeint(lorenz, L3, t, args=paras)

L4 = [1.0, 1.01, 1.0]
track4 = odeint(lorenz, L4, t, args=(10.0, 10.0, 8.0/3.0))

fig = plt.figure(figsize=(20, 10))
ax = fig.gca(projection="3d")
ax.plot(track1[:,0], track1[:,1], track1[:,2], color='r',lw=0.5) # 绘制轨迹 1
ax.plot(track3[:,0], track3[:,1], track3[:,2], color='g',lw=0.5) # 绘制轨迹 2plt1.set_title("Lorenz Attractor 1 by odeint")
ax.set_title('Lorenz Strange Attractor 1')

fig = plt.figure(figsize=(20, 10))
ax = fig.gca(projection="3d")
ax.plot(track2[:,0], track2[:,1], track2[:,2], color='r',lw=0.5) # 绘制轨迹 1
ax.plot(track4[:,0], track4[:,1], track4[:,2], color='g',lw=0.5) # 绘制轨迹 2plt1.set_title("Lorenz Attractor 1 by odeint")
ax.set_title('Lorenz Strange Attractor 2')
```

_P.S._ 为了巩固新知识，原文的注释也差不多照打了一遍。动图没画出来。

```python
a = track1[:,0] - track3[:,0]
b = track1[:,1] - track3[:,1]
c = track1[:,2] - track3[:,2]
d = track2[:,0] - track4[:,0]
e = track2[:,1] - track4[:,1]
f = track2[:,2] - track4[:,2]

distance1 = np.sqrt(a*a+b*b+c*c)
ln1 = np.log(distance1/0.01)
distance2 = np.sqrt(d*d+e*e+f*f)
ln2 = np.log(distance2/0.01)

fig,ax4 = plt.subplots(figsize=(10,5))
ax4.plot(t,ln1,color='purple')
ax.set_title('Figure 1')

fig,ax5 = plt.subplots(figsize=(10,5))
ax5.plot(t,ln2,color='purple')
ax.set_title('Figure 2')
```
